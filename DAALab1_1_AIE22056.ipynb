```python
import time
import matplotlib.pyplot as plt
import random
```

```python
# 1.Sum of first N natural numbers using iterative and recursive algorithms
def iter_Sum(N):
    sum = 0
    for i in range(1, N+1):
        sum += i
    return sum

def Recursive_Sum(N):
    if N == 1:
        return 1
    else:
        return N + Recursive_Sum(N-1)

# Time comparison
N_value = [10, 50, 100, 500]
iter_time = []
recs_time = []

for N in N_value:
    start_time = time.time()
    iter_Sum(N)
    iter_time.append(time.time() - start_time)

    start_time = time.time()
    Recursive_Sum(N)
    recs_time.append(time.time() - start_time)

print("Iterative Time: ",iter_time)
print("\nRecursive Time: ",recs_time,"\n")

# Plotting
plt.plot(N_value, iter_time, label='Iterative')
plt.plot(N_value, recs_time, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (seconds)')
plt.title('Time Comparison of Iterative vs Recursive Sum')
plt.legend()
plt.show()
```

```python
# 2.Linear and binary search for an array of 10000 elements
def LinearSearch(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1

def BinarySearch(A, key):
    low = 0
    high = len(A) - 1
    while low <= high:
        mid = (low + high) // 2
        if A[mid] == key:
            return mid
        elif A[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Generate random array
A = [random.randint(1, 1000) for _ in range(10000)]

# Perform searches
keys_val = [random.randint(1, 1000) for _ in range(5)]
lTime = []
bTime = []

for key in keys_val:
    start_time = time.time()
    LinearSearch(A, key)
    lTime.append(time.time() - start_time)

    start_time = time.time()
    BinarySearch(A, key)
    bTime.append(time.time() - start_time)

# Plotting
plt.plot(range(1, 6), lTime, label='Linear Search')
plt.plot(range(1, 6), bTime, label='Binary Search')
plt.xlabel('Search Elements')
plt.ylabel('Time (seconds)')
plt.title('Time Comparison of Linear Search vs Binary Search')
plt.legend()
plt.show()
```

```python
#3.Recursive function to convert a string of digits into an integer
def str2int(s):
    if len(s) == 1:
        return int(s)
    else:
        #place of number unit,tens,etc. and add previous
        return 10**(len(s)-1) * int(s[0]) + str2int(s[1:])

# Example usage
s = input("Enter a number: ")
print(type(s))
print(str2int(s))
print(type(str2int(s)))
```

```python
#4.Recursive function to reverse a string
def reverse_str(s):
    if len(s) == 0:
        return s
    else:
        return reverse_str(s[1:]) + s[0]

s = input("Enter string: ")
print(reverse_str(s))
```

```python
#5.Recursive function to determine if a string is a palindrome
def is_palindrome(s):
    if len(s) <= 1:
        return True
    elif s[0] != s[-1]:
        return False
    else:
        return is_palindrome(s[1:-1])

# Example usage
s = input("Enter a word: ")
print(is_palindrome(s))
```
